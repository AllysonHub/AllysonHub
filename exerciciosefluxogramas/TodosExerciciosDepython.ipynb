##Calcular horas trabalhadas + 70%

image.png

```python
h = float(input('Digite as Horas Trabalhadas: '))
m = float(input("Digite os Minutos Trabalhados: "))
g = float(input("Digite Quanto voce ganha Por Hora: "))

# Processamento
total_horas = h + m / 60
salario_base = total_horas * g
salario_extra = salario_base * 0.70
salario_total = salario_base + salario_extra

# Saída
print('Salário total incluindo horas extras: R$', salario_total)
```
